```python
label = gold

for z in range(0,1):

    # fill segmentation
    label_zeros = _metrics.Util.threshold(label[z], 0)
    label_filled = _metrics.Util.fill(label[z], label_zeros.astype(np.bool))
    label_filled_relabeled = skimage.measure.label(label_filled).astype(np.uint64)

    print 'Working on z', z

    labels = range(len(_metrics.Util.get_histogram(label_filled_relabeled)))#[1:] ### remove!!
    for l in labels:
        
        # these are all the labels of the current slice
        
        # let's check how many real neighbors this label has
        neighbors = _metrics.MergeError.grab_neighbors(label_filled_relabeled, l)
        neighbors_count = len(neighbors)
        
        print 'label',l,'has',neighbors_count,'labels'
        
        
```

```python
#
# correct patches
#

import time

t0 = time.time()
patches = 0
data = []

NO_PATCHES = 100
PATCH_BYTES = 75*75
p_image = np.zeros((NO_PATCHES, PATCH_BYTES),dtype=np.uint8)
p_prob = np.zeros((NO_PATCHES, PATCH_BYTES),dtype=np.uint8)
p_label1 = np.zeros((NO_PATCHES, PATCH_BYTES),dtype=np.uint8)
p_label2 = np.zeros((NO_PATCHES, PATCH_BYTES),dtype=np.uint8)
p_overlap = np.zeros((NO_PATCHES, PATCH_BYTES),dtype=np.uint8)


for s in _metrics.MergeError.generate(images[0:1], probs[0:1], gold[0:1], thumb=False, rotate=True):
#for s in _metrics.SplitError.generate(images, gold, 10, thumb=False, rotate=True):    

    #data.append(s)

#     if s._meta['label1'].astype(np.uint8).max() == 0:
#         print 'wrong egg'
#         continue
    
    # now we have a correct patch
#     cv2.imwrite(TRAINING_PATH+str(splits)+'_image.tif', s._meta['image'])
#     cv2.imwrite(TRAINING_PATH+str(splits)+'_prob.tif', s._meta['prob'])
#     cv2.imwrite(TRAINING_PATH+str(splits)+'_label1.tif', img_as_ubyte(s._meta['label1']))
#     cv2.imwrite(TRAINING_PATH+str(splits)+'_label2.tif', img_as_ubyte(s._meta['label2']))
#     cv2.imwrite(TRAINING_PATH+str(splits)+'_overlap.tif', img_as_ubyte(s._meta['overlap']))

    p_image[patches] = s._meta['image'].ravel()
    p_prob[patches] = s._meta['image'].ravel()
    p_label1[patches] = s._meta['image'].ravel()
    p_label2[patches] = s._meta['image'].ravel()
    p_overlap[patches] = s._meta['image'].ravel()    

    t1 = time.time()
    total = t1-t0

    patches += 1
    if patches % 1000 == 0:
        print 'Another 1000 generated after',total,'seconds'    
    
#     if total > 100:
#         break

    if patches >= NO_PATCHES:
        break


print total, 'seconds for', patches, 'patches'
```
