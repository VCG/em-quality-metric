```python
import cv2
import glob
import numpy as np
import os
import tifffile as tif

from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
DATA_PATH = '/Volumes/DATA1/EMQM_DATA/ac3x75/'
IMAGE_PATH = os.path.join(DATA_PATH,'input/')
GOLD_PATH = os.path.join(DATA_PATH,'gold/')
RHOANA_PATH = os.path.join(DATA_PATH, 'rhoana/')

def get_section(n, path):
    return glob.glob(path+'*.tif')[0]
```

```python
SECTION = 0
```

```python
image = cv2.imread(get_section(SECTION, IMAGE_PATH),0)
imshow(image, cmap='gray')
```

```python
gold = cv2.imread(get_section(SECTION, GOLD_PATH),0)
imshow(gold, cmap='gray')
```

```python
rhoana = tif.imread(get_section(SECTION, RHOANA_PATH))
imshow(rhoana.astype(np.float), cmap='gray')
```

```python
#
# entropy
#
from skimage.filters.rank import entropy
from skimage.morphology import disk
```

```python
image_entropy = entropy(image, disk(100))
imshow(image_entropy)
```

```python
gold_entropy = entropy(gold, disk(100))
imshow(gold_entropy)
```

```python
rhoana_entropy = entropy(rhoana.astype(np.uint8), disk(100))
imshow(rhoana_entropy)
```
