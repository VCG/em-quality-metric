```python
from scipy.stats import describe

sample_rates = [1,2,3,10,-1]
center_minus_means_per_samplerate = []

PRINT_ALL = False



for sample_rate in sample_rates:



    labels = range(1,len(_metrics.Util.get_histogram(split_gold_normalized_framed.astype(np.uint64))))

    center_minus_means = []


    for l in labels:

        neighbors = grab_neighbors(split_gold_normalized_framed, l)

        for n in neighbors:

            if n == 0:
                continue



            p_sample = grab_patch(image, prob, split_gold_normalized_framed, l, n, sample_rate=sample_rate)
    #         if np.any(p_sample, None):
    #             print 'ERROR'
    #             break


            if p_sample and PRINT_ALL:
                print '-'*80
                print 'checking', l, n



            results = []

            for i,p in enumerate(p_sample):

                if i == 0:
                    center_value = test_patch(val_fn, p)

                if p['image'].shape != (75,75):
                    continue            

                results.append(test_patch(val_fn, p))

            if len(results) > 0:
                d_results = describe(results)
                center_mean = abs(center_value - d_results[2])
                center_minus_means.append(center_mean)            
                if PRINT_ALL:
                    print 'N', d_results[0]
                    print 'center', round(center_value,3)
                    print 'minmax', round(d_results[1][0],3),round(d_results[1][1],3)
                    print 'mean', round(d_results[2],3)

                    print 'abs(center-mean)', round(center_mean, 3)
                    print 'variance', round(d_results[3],3)    
        #             print 'skewness', round(d_results[3],3)
                    print 'normal distribution (if 0)', round(d_results[4],3)

            
    center_minus_means_per_samplerate.append(center_minus_means)
```
