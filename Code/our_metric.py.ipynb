```python
def calculate_metrics(array, PATCH_SIZE_YX):
    '''
    Needs a 3D array Z,Y,X.
    '''

    output = []
    
    w = array.shape[2]
    h = array.shape[1]
    
    offset_x = 0
    offset_y = 0
    
    no_patches_y = w / PATCH_SIZE_YX[1]
    no_patches_x = w / PATCH_SIZE_YX[0]

    for z in range(array.shape[0]):
        
        x_measures = []
        
        for px in range(no_patches_x):
            y_measures = []
            
            for py in range(no_patches_y):
            
                patch_measure = {}
            
                patch = array[z:z+1, py*PATCH_SIZE_YX[0]:py*PATCH_SIZE_YX[0]+PATCH_SIZE_YX[0], px*PATCH_SIZE_YX[1]:px*PATCH_SIZE_YX[1]+PATCH_SIZE_YX[1]]

                patch_measure['Patch'] = (z,py,px)
                patch_measure['SizeDistributionMeasure'] = _metrics.SizeDistributionMeasure.apply(patch, False)
                
                y_measures.append(patch_measure)
                
            x_measures.append(y_measures)
                
        output.append(x_measures)

    return output

def compare(measures1, measures2):
    
    print_info = True
    
    stats = {}
    
    for z in range(len(measures1)):
        for x in range(len(measures1[z])):
            for y in range(len(measures1[z][x])):
                patch1 = measures1[z][x][y]['Patch']
                patch2 = measures2[z][x][y]['Patch']
                
                comparison = str(patch1)+' '
                
                for size_distribution_measure in measures1[z][x][y]['SizeDistributionMeasure']:

                    if print_info:
                        #print size_distribution_measure
                        stats[size_distribution_measure] = {'G':0,'L':0,'E':0}
                    
                    # now compare
                    if (measures1[z][x][y]['SizeDistributionMeasure'][size_distribution_measure] > measures2[z][x][y]['SizeDistributionMeasure'][size_distribution_measure]):
                        comparison += 'G'
                        stats[size_distribution_measure]['G'] += 1
                    elif (measures1[z][x][y]['SizeDistributionMeasure'][size_distribution_measure] < measures2[z][x][y]['SizeDistributionMeasure'][size_distribution_measure]):
                        comparison += 'L'
                        stats[size_distribution_measure]['L'] += 1                        
                    elif (measures1[z][x][y]['SizeDistributionMeasure'][size_distribution_measure] == measures2[z][x][y]['SizeDistributionMeasure'][size_distribution_measure]):
                        comparison += 'E'
                        stats[size_distribution_measure]['E'] += 1                        

                print_info = False
                    
                #print comparison
    return stats
                
    
def add_merge_errors(array, merges):
    
    # get the list of segments
    segment_sizes = _metrics.Util.get_histogram(array.astype(np.uint64))
    
    segments = np.arange(0,len(segment_sizes))
    merge_table = np.arange(0,len(segment_sizes))
    
    for m in range(merges):
        segment1 = np.random.choice(segments)
        
        # detect if we already have a merge for this segment
        while merge_table[segment1] != segment1 or segment1 == 0:
            segment1 = np.random.choice(segments)
        
        segment2 = segment1
        while segment1 == segment2:
            segment2 = np.random.choice(segments)
            
        # now we have two random segments
        merge_table[segment1] = segment2
    
    print merge_table    
    array = merge_table[array]
    
    return array
    
    
def fit_segment_to_patch(array, shape_threshold):
    
    for z in range(array.shape[0]):
        
        normalized_section = _metrics.Util.normalize_labels(array[z,:,:])[0]
        
        section_bbox = mh.labeled.bbox(normalized_section)
        no_segments = len(section_bbox)
        counter = 0
        
        for s in section_bbox:
            
            #w = s[5]-s[4]
            #h = s[3]-s[2]
            w = s[3]-s[2]
            h = s[1]-s[0]
            
            if w == 0 or h == 0:
                continue
            
            if w <= shape_threshold[1] and h <= shape_threshold[0]:
                # this segment fits
                counter += 1
                
        print 'Slice', str(z)+':',counter,'/',no_segments,'(',(float(counter) / no_segments)*100,'%)','fit.' 
        
```

```python
from scipy import ndimage as nd

def fill(data, invalid=None):
    """
    Replace the value of invalid 'data' cells (indicated by 'invalid') 
    by the value of the nearest valid data cell

    Input:
        data:    numpy array of any dimension
        invalid: a binary array of same shape as 'data'. 
                 data value are replaced where invalid is True
                 If None (default), use: invalid  = np.isnan(data)

    Output: 
        Return a filled array. 
    """    
    if invalid is None: invalid = np.isnan(data)

    ind = nd.distance_transform_edt(invalid, 
                                    return_distances=False, 
                                    return_indices=True)
    return data[tuple(ind)]
```
