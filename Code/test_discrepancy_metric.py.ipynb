```python
import _metrics


import cv2
import glob
import numpy as np
import mahotas as mh
import os
import tifffile as tif

from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
DATA_PATH = '/Volumes/DATA1/EMQM_DATA/ac3x75/'
IMAGE_PATH = os.path.join(DATA_PATH,'input/')
GOLD_PATH = os.path.join(DATA_PATH,'gold/')
RHOANA_PATH = os.path.join(DATA_PATH, 'rhoana/')

def get_section(n, path):
    return glob.glob(path+'*.tif')[0]
```

```python
SECTION = 0
```

```python
image = cv2.imread(get_section(SECTION, IMAGE_PATH),0)
imshow(image, cmap='gray')
```

```python
gold = tif.imread(get_section(SECTION, GOLD_PATH),0)
imshow(gold, cmap='gray')
```

```python
rhoana = tif.imread(get_section(SECTION, RHOANA_PATH))
imshow(rhoana.astype(np.float), cmap='gray')
```

```python
rhoana_largest = _metrics.Util.threshold(rhoana, _metrics.Util.get_largest_label(rhoana))
imshow(rhoana_largest)
```

```python
gold_largest = _metrics.Util.threshold(gold, _metrics.Util.get_largest_label(gold))
imshow(gold_largest)
```

```python
rhoana_global_measure = 0
rhoana_normalized = _metrics.Util.normalize_labels(rhoana)[0]
rhoana_labels = _metrics.Util.get_histogram(rhoana_normalized.astype(np.uint16))

for l,k in enumerate(rhoana_labels):
    if l == 0:
        continue
    thresholded_rhoana = _metrics.Util.threshold(rhoana_normalized, l)
    rhoana_global_measure += _metrics.DiscrepancyMeasure.apply2D(thresholded_rhoana, image)
rhoana_global_measure /= len(rhoana_labels)
print rhoana_global_measure
```

```python
gold_global_measure = 0
gold_normalized = _metrics.Util.normalize_labels(gold)[0]
gold_labels = _metrics.Util.get_histogram(gold_normalized.astype(np.uint16))

for l,k in enumerate(gold_labels):
    if l == 0:
        continue    
    thresholded_gold = _metrics.Util.threshold(gold_normalized, l)
    gold_global_measure += _metrics.DiscrepancyMeasure.apply2D(thresholded_gold, image)
gold_global_measure /= len(gold_labels)
print gold_global_measure
```

```python
rhoana_normalized = _metrics.Util.normalize_labels(rhoana)[0]
thresholded_rhoana = _metrics.Util.threshold(rhoana_normalized, 1)
imshow(thresholded_rhoana, cmap='gray')
print _metrics.DiscrepancyMeasure.apply(thresholded_rhoana, image)
```
