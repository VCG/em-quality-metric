```python
%load_ext autoreload
%autoreload 2

from cnn import CNN
from util import Util
from patch import Patch
from fixer import Fixer
from uglify import Uglify

import mahotas as mh
import numpy as np

from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cnn = CNN('8b33e516-2311-4892-82f9-4e6079c261f8', 'patches_4th_10k')
```

```python
SLICE = 70
input_image, input_prob, input_gold, input_rhoana = Util.read_section(SLICE, keep_zeros=False) 
```

```python
framed_gold = Util.frame_image(input_gold, shape=(200,200))
```

```python
Util.view(framed_gold, large=True)
```

```python
hist = Util.get_histogram(framed_gold.astype(np.uint64))
labels = range(len(hist))
np.random.shuffle(labels)

min_pixels = 100

for l in labels[:10]:

    if l == 0:
        continue

    if len(framed_gold[framed_gold == l]) < min_pixels:
        continue
        
    neighbors = Util.grab_neighbors(framed_gold, l)
    np.random.shuffle(neighbors)
    
    for n in neighbors:
    
        if n == 0:
            continue
            
        if len(framed_gold[framed_gold == n]) < min_pixels:
            continue

        break
    
    print 'merging', l, n



    cropped_image, cropped_prob, cropped_segmentation, cropped_binary, bbox = Uglify.merge_label(input_image,
                                                                                           input_prob,
                                                                                           framed_gold,
                                                                                           l,
                                                                                           n,
                                                                                           crop=True)

    borders, best_border_image, result = Fixer.fix_single_merge(cnn, cropped_image, cropped_prob, cropped_binary)

    Util.view_labels(framed_gold, [l, n])
    Util.view_labels(cropped_segmentation, [l])
    Util.view(borders, large=True, color=False)
    Util.view(result, large=True, crop=True)




```

```python
label1 = framed_gold[600,600]
label2 = framed_gold[750,600]




cropped_image, cropped_prob, cropped_segmentation, cropped_binary, bbox = Uglify.merge_label(input_image,
                                                                                       input_prob,
                                                                                       framed_gold,
                                                                                       label1,
                                                                                       label2,
                                                                                       crop=False)

borders, best_border_image, result = Fixer.fix_single_merge(cnn, cropped_image, cropped_prob, cropped_binary)

Util.view_labels(framed_gold, [label1, label2])
Util.view_labels(cropped_segmentation, [label1])
Util.view(result, large=True, crop=True)
```
