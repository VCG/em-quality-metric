```python
%load_ext autoreload
%autoreload 2

from cnn import CNN
from util import Util
from patch import Patch
from fixer import Fixer
from uglify import Uglify

import mahotas as mh
import numpy as np

from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cnn = CNN('8b33e516-2311-4892-82f9-4e6079c261f8', 'patches_4th_10k')
```

```python
SLICE = 70
# zeros are ignored during patch generation!
input_image, input_prob, input_gold, input_rhoana = Util.read_section(SLICE, keep_zeros=True) 
```

```python
framed_gold = Util.frame_image(input_gold)
ugly_gold = Uglify.split(input_image, input_prob, framed_gold, n=2)

print 'Labels before:', len(Util.get_histogram(framed_gold))
print 'Labels after:', len(Util.get_histogram(ugly_gold))
print 'VI after uglifying:', Util.vi(ugly_gold, framed_gold)
```

```python
Util.view(ugly_gold, large=True)
```

```python
vi_s, merge_pairs, surenesses = Fixer.splits(cnn, input_image, input_prob, ugly_gold, framed_gold)
```

```python
Util.merge_steps(ugly_gold, merge_pairs, best=vi_s.index(np.min(vi_s)))
```

```python
rhoana_vi_s, rhoana_merge_pairs, rhoana_surenesses = Fixer.splits(cnn, input_image, input_prob, input_rhoana, framed_gold)
```

```python
Util.view(input_rhoana, large=True)
```

```python
Util.merge_steps(input_rhoana, rhoana_merge_pairs, best=rhoana_vi_s.index(np.min(rhoana_vi_s)))
```

```python
rhoana_without_zeros = np.array(input_rhoana)
rhoana_without_zeros[framed_gold == 0] = 0
Util.view(rhoana_without_zeros, large=True)
```

```python
rhoana_without_zeros_vi_s, rhoana_without_zeros_merge_pairs, rhoana_without_zeros_surenesses = Fixer.splits(cnn, input_image, input_prob, rhoana_without_zeros, framed_gold)
```

```python
Util.merge_steps(rhoana_without_zeros, rhoana_without_zeros_merge_pairs, best=rhoana_without_zeros_vi_s.index(np.min(rhoana_without_zeros_vi_s)))
```
