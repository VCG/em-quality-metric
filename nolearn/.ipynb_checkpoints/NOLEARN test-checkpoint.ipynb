```python
from __future__ import division, print_function

IMAGE_HEIGHT = 28
IMAGE_WIDTH = 28

import scipy
import theano
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns
%matplotlib inline

from sklearn.datasets import fetch_mldata
from sklearn.cross_validation import train_test_split

from nolearn.lasagne import NeuralNet
from lasagne import layers
from lasagne import nonlinearities
from lasagne.updates import nesterov_momentum
```

```python
mnist = fetch_mldata('MNIST original')
X = mnist.data.astype(np.float32) / 255
y = mnist.target.astype(np.int32)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

X_train = X_train.reshape(-1, 1, IMAGE_HEIGHT, IMAGE_WIDTH)
X_test = X_test.reshape(-1, 1, IMAGE_HEIGHT, IMAGE_WIDTH)

print(X_train.shape, X_test.shape)
```

```python
clf = NeuralNet(
    layers=[
        ('input', layers.InputLayer),

        ('conv1', layers.Conv2DLayer),
        ('pool1', layers.MaxPool2DLayer),
        ('dropout1', layers.DropoutLayer),

        ('conv2', layers.Conv2DLayer),
        ('pool2', layers.MaxPool2DLayer),
        ('dropout2', layers.DropoutLayer),

        ('hidden3', layers.DenseLayer),
        ('dropout3', layers.DropoutLayer),
        ('output', layers.DenseLayer),
    ],

    input_shape=(None, 1, IMAGE_HEIGHT, IMAGE_WIDTH),
    output_num_units=10,
    output_nonlinearity=nonlinearities.softmax,

    conv1_filter_size=(5,5), conv1_num_filters=32,
    pool1_pool_size=(2,2),
    dropout1_p=0.1,

    conv2_filter_size=(5,5), conv2_num_filters=32,
    pool2_pool_size=(2,2),
    dropout2_p=0.2,

    hidden3_num_units=256,
    dropout3_p=0.5,

    update=nesterov_momentum,
    update_learning_rate=0.01,
    update_momentum=0.9,

    max_epochs=50,
    verbose=True
)

clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf_layers = clf.get_all_layers()
```

```python
layer = clf_layers[1]
W = layer.W.get_value()
b = layer.b.get_value()
f = [w + bb for w, bb in zip(W, b)]

gs = gridspec.GridSpec(6, 6)
for i in range(layer.num_filters):
    g = gs[i]
    ax = plt.subplot(g)
    ax.grid()
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(f[i][0])
```

```python
output = layer.get_output_for(X_test[:1]).eval()[0]
gs = gridspec.GridSpec(6, 6)
for i in range(layer.num_filters):
    g = gs[i]
    ax = plt.subplot(g)
    ax.grid()
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(output[i])
```
